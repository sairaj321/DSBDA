```python
# 1 pie chart using Matplotib<<<<<<<<<
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Plot
#plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
# 2 Bar chart using matplotlib
categories = ['Category 1', 'Category 2', 'Category 3']
values1 = [20, 35, 30]
values2 = [25, 32, 28]
# Plot
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(categories))
plt.bar(index, values1, bar_width, label='Group 1')
#plt.bar([i + bar_width for i in index], values2, bar_width, label='Group 2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
# 3 Box plot using matplotlib
# Data
np.random.seed(10)
data = np.random.normal(loc=0, scale=1, size=(100, 5))

# Plot

plt.boxplot(data, notch=True, vert=True, patch_artist=True)
plt.title('Box Plot')
plt.show()
```

```python
#Bar chart using SNS
import pandas as pd
import seaborn as sns
df=sns.load_dataset('titanic')
df_pivot=pd.pivot_table(df,values="fare",index="who",columns="class",aggfunc=np.mean)
ax=df_pivot.plot(kind="bar",alpha=0.5)
plt.title('Bar ')
plt.xlabel('X axis title')
plt.ylabel('Y axis title')
plt.show()
```

```python
#line chart using SNS
df=sns.load_dataset("iris")
df=df.groupby('sepal_length')['sepal_width'].sum().to_frame().reset_index()
plt.plot(df['sepal_length'],df['sepal_length'])
plt.title('line chart')
plt.xlabel('X axis title')
plt.ylabel('Y axis title')
plt.show()
```

```python
#histogram using matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(0)
data1 = np.random.normal(loc=0, scale=1, size=1000)
data2 = np.random.normal(loc=2, scale=1.5, size=1000)

# Plot
plt.figure(figsize=(10, 6))
plt.hist([data1, data2], bins=30, color=['skyblue', 'salmon'], edgecolor='black', alpha=0.7, label=['Data 1', 'Data 2'])
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.show()
```

```python
# scatter chart using matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
sizes = np.random.randint(10, 200, 100)
colors = np.random.rand(100)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.colorbar(label='Color intensity')
plt.grid(True)
plt.show()
```

```python
#line chart using matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Sin(x)', linestyle='-', linewidth=2, color='blue')
plt.plot(x, y2, label='Cos(x)', linestyle='--', linewidth=2, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Graph')
plt.legend()
plt.grid(True)
plt.show()
```
