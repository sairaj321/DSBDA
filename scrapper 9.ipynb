```python
import requests
import bs4
```

```python
request1=requests.get('https://www.flipkart.com/apple-iphone-15-black-128-gb/p/itm6ac6485515ae4')
request1
```

```python
request1.content
```

```python
soup=bs4.BeautifulSoup(request1.text)
soup
```

```python
#fetching Reviews/ comments
reviews=soup.findAll('div',{'class':'ZmyHeo'});
for review in reviews:
    print(review.get_text()+"\n\n")
```

```python
#average Overall Rating 
rating =soup.find('div',{'class':'ipqd2A'}).get_text();
print(rating)
```

```python
#Individual Rating 
individual_rating = soup.findAll('div',{'class':'XQDdHH Ga3i8K'});
for indi_rating in individual_rating:
    print(indi_rating.get_text() +"\n")
```

```python
#tags
tags=soup.find('span',{'class':'_6EBuvT OHdLLG'}).get_text();
tags
```

```python
#customer name
customer_name = soup.findAll('p',{'class':'_2NsDsF AwS1CA'});
for cust_name in customer_name:
    print(cust_name.get_text() +"\n")
```
